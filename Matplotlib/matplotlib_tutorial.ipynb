```python
x = [1,2,3,4]
dsa_marks = [90, 92, 96, 98]
ml_marks = [93, 95, 98, 100]

plt.plot(x, dsa_marks, label = 'dsa marks', color='m', linewidth=2, marker='*', linestyle='--', markersize = 20)
plt.plot(x, ml_marks, label = 'ml marks')

plt.title("Marks corresponding to no of hrs studied", 
          fontdict={'fontname': 'comic sans ms', 'fontweight': 20, 'fontsize' : 20,  'color': 'blue'})

plt.xlabel("No of hrs studied", 
           fontdict={'fontname': 'times new roman', 'fontweight': 5, 'fontsize' : 10, 'color': 'red'})
plt.ylabel("Marks",
          fontdict={'fontname': 'times new roman', 'fontweight': 5, 'fontsize' : 10,'color': 'red'})

plt.xticks([1,2,3,4])
plt.yticks([90, 92,94,96,98,100])

plt.legend(loc="lower right")

plt.savefig("marks.png", dpi=1000, transparent = 1)

plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.figure()

plt.subplot(3, 1, 1)
plt.plot(t, t, 'r--')
plt.xlabel('t values')
plt.ylabel('output values')
plt.title('Linear Subplot')

plt.subplot(3, 1, 2)
plt.plot(t, t**2, 'bs')
plt.xlabel('t values')
plt.ylabel('output values')
plt.title('Quadratic Subplot')

plt.subplot(3, 1, 3)
plt.plot(t, t**3, 'g^')
plt.xlabel('t values')
plt.ylabel('output values')
plt.title('Cubic Subplot')

plt.suptitle('Different functions of t')
plt.tight_layout()
plt.xlabel('t values')
plt.ylabel('output values')

plt.show()
```

```python
# axes - x axis, y axis, labels, tickers

x = np.linspace(1, 10, 100)

fig, axes = plt.subplots(2, 3)

axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title('Sin func')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')

axes[0,1].plot(x, np.tan(x))
axes[0,1].set_title('Tan func')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')

axes[0,2].plot(x, np.exp(x))
axes[0,2].set_title('Exp func')
axes[0,2].set_xlabel('x')
axes[0,2].set_ylabel('y')

axes[1,0].plot(x, np.sqrt(x))
axes[1,0].set_title('Sqrt func')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')

axes[1,1].plot(x, np.log(x))
axes[1,1].set_title('Log func')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')

axes[1,2].plot(x, x)
axes[1,2].set_title('Linear func')
axes[1,2].set_xlabel('x')
axes[1,2].set_ylabel('y')


plt.tight_layout()


```

### Scatter Plot

```python
x = np.random.rand(100) * 10
y = np.random.rand(100) * 10

plt.scatter(x,y, marker='D', edgecolors='black', linewidths=2, alpha=0.3, s = 50, label= 'random')

plt.legend(loc="lower right")
plt.grid(True)

plt.xlabel('x')
plt.ylabel('y')
plt.title('random scattered plot')
plt.xticks([0,5,10])
plt.yticks([0,5,10])

plt.show()
```

### Bar Plot

```python
courses = ['GenAI' , 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]

plt.bar(courses, students, width=0.5, align='center', edgecolor='black', linewidth=3, color = 'red', hatch='*')
plt.xlabel('Courses')
plt.ylabel('No of students')
plt.title('No of students enrolled in courses')

plt.grid(True)

plt.show()
```

### Histogram Plot

```python
ages = [34, 23, 78, 54, 67, 45, 21, 98, 54, 76]

sorted_ages = np.sort(ages)
print(sorted_ages)

plt.hist(ages, bins = 5, orientation='horizontal', linewidth=2, edgecolor='black', color='red')
plt.xlabel('Count')
plt.ylabel('Ages')
plt.title('Ages Plot')

plt.show()
```

### Pie Plot

```python
courses = ['GenAI' , 'HLD', 'HHLD', 'LLD', 'DSA', 'CPP']
students = [2500, 3000, 3200, 2800, 2700, 1000]
explode = [0.2, 0, 0, 0, 0, 0]
pie_colors = ['red', 'blue', 'magenta', 'yellow', 'orange', 'green']

plt.pie(students, labels=courses, shadow=True, explode=explode, colors=pie_colors, autopct="%.2f%%")
plt.legend()
plt.show()
```

### 3D Scatter Plot

```python
ax = plt.axes(projection="3d")

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_title('3D Scattered Plot')

plt.show()
```

### 3D Plot

```python
ax = plt.axes(projection="3d")

x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


ax.plot(x, y, z)

plt.show()
```

### 3D Bar PLot

```python
ax = plt.axes(projection="3d")

x = np.random.randint(1, 10, 5)
y = np.random.randint(1, 10, 5)
z = np.zeros(5)

dx = np.ones(5) * 0.5
dy = np.ones(5)
dz = np.random.randint(1, 10, 5)

print(x, y, z, dx, dy, dz)

ax.bar3d(x, y, z, dx, dy, dz)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


plt.show()
```

```python
ax = plt.axes(projection="3d")

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, cmap='plasma')

plt.show()
```
